```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

evenly spaced values

```python
x = np.linspace(10,20,5)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

SINE WAVE

```python
plt.plot(x,y)
plt.show()
```

COS WAVE

```python
plt.plot(x,z)
plt.show()
```

ADDING TITLE x- axis & y- axis

```python
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine value")
plt.title("sine wave")
plt.show()
```

PARABOLA

```python
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

BAR DIAGRAM

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['english','french','japanese','german','tamil']
people=[100,60,40,20,500]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

PIE CHART

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['english','french','japanese','german','tamil']
people=[100,60,40,20,500]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

SCATTER GRAPH

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color='g')
ax.scatter(x,z,color='r')
plt.show()
```

3D PROJECTION

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Reds')
plt.show()
```
